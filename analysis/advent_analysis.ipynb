```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("CoffeeAdvent_CR/analysis/advent_data.csv", header = 0)
```

```python
# Clean dataset
#Convert strings to lowercase
string_columns = df.select_dtypes(include=['object']).columns
df[string_columns] = df[string_columns].apply(lambda x: x.str.lower())

# Remove last rows
df = df.iloc[:26]

#print(df.head(10))
#print(df.columns)
#print(df.head(10))


# Format onyx_notes
df['onyx_notes'] = df['onyx_notes'].str.replace('\s*\|\s*', ', ', regex=True).str.strip()

```

```python
# Correlations

shervin_caitlin = df['shervin_rating'].corr(df['caitlin_rating'])
print(f"Shervin and Caitlin agreement: {shervin_caitlin}")

shervin_cupscore = df['shervin_rating'].corr(df['cup_score'])
print(f"Shervin and Cup Score agreement: {shervin_cupscore}")

#re-run with 50 removed
caitlin_cupscore = df['caitlin_rating'].corr(df['cup_score'])
print(f"Caitlin and Cup Score agreement: {caitlin_cupscore}")

shervin_agtron = df['shervin_rating'].corr(df['agtron'])
print(f"Shervin and agtron agreement: {shervin_agtron}")

caitlin_agtron = df['caitlin_rating'].corr(df['agtron'])
print(f"Caitlin and agtron agreement: {caitlin_agtron}")

shervin_green = df['shervin_rating'].corr(df['onyx_green_price'])
print(f"Shervin and green price agreement: {shervin_green}")

caitlin_green = df['caitlin_rating'].corr(df['onyx_green_price'])
print(f"Caitlin and green agreement: {caitlin_green}")

shervin_box = df['shervin_rating'].corr(df['onyx_box_price'])
print(f"Shervin and box price agreement: {shervin_box}")

caitlin_box = df['caitlin_rating'].corr(df['onyx_box_price'])
print(f"Caitlin and box agreement: {caitlin_box}")
```

```python
# Shervin Caitlin Correlation Plot

plt.scatter(df['shervin_rating'], df['caitlin_rating'])
plt.xlabel('Shervin Rating')
plt.ylabel('Caitlin Rating')
plt.title('Correlation between Shervin and Caitlin Ratings')
plt.show()
```

```python
# Shervin Box Price Correlation Plot

plt.scatter(df['shervin_rating'], df['onyx_box_price'])
plt.xlabel('Shervin Rating')
plt.ylabel('Onyx Box Price')
plt.title('Correlation between Shervin Ratings and Box Price')
plt.show()
```
